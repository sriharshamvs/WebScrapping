```python
import pandas as pd
```

```python
path = "ElectionResults.csv"
ElectionResults = pd.read_csv(path, error_bad_lines=False, warn_bad_lines=False)
ElectionResults.head(2)
```

```python
ElectionResults = ElectionResults.drop(['O.S.N.'], axis=1)
ElectionResults.head(5)
```

```python
ElectionResults_ST_UT = ElectionResults.groupby('State')
ElectionResults_ST_UT.get_group('Telangana').head(5)
# ElectionResults_ST_UT
```
